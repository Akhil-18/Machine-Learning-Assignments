# A. Linear Algebra
Linear Algebra is a branch of mathematics which deals with linear equations. It consists of several mathematical objects: <br> 
**Scalars**: A scalar is a single number. <br>
**Vectors**: A vector is an array of numbers. <br> 
**Matrices**: A matix is a 2-D array of numbers, where element is identified by two indices representing rows and columns.<br>
**Tensors**: An array of numbers in a grid with variable number of axes. <br>
Another important term is **transpose**. A transpose of a matrix is mirror image of matrix along its main diagonal. 

**Multiplying Matrices and Vectors** <br>
The **matrix product** of A and B is a third matrix **C. <br> We can write: C=AB** <br>


\begin{matrix}
{AB}]_{i,j} = a_{i,1}b_{1,j} + a_{i,2}b_{2,j} + \cdots + a_{i,n}b_{n,j} = \sum_{r=1}^n a_{i,r}b_{r,j}
\end{matrix}

The dot product of two vector x and y of same dimensionality is the matrix product 
\begin{align}
\mathbf{\color{red}a} \cdot \mathbf{\color{blue}b} = \mathbf{\color{red}a}\mathbf{\color{blue}b}^\top\end{align}

1. Matrix multiplication is distributive.
2. Matrix multiplication is not commutative. 
3. Matrix multiplication is associative.

**Identiy and Inverse Matrices**

An identity matrix is that which doesn't change any vector when multiplied by that vector. <br>
\begin{align}(I_{n})_{ij}=\delta _{ij}. \end{align}

The matrix inverse of A is denoted by A<sup>−1 </sup> and is defined as a matrix 
AA<sup>-1 </sup>=I<sub>n</sub>

**Linear Dependence and Span**
For A<sup>-1</sup> to exists, the linear equation Ax=b must have exactly one solution for every value of b. It is also possible for system to have exactly one solution or infinitely many solution but there will be no such case as more than one but less than infinitely many solutions for a particular value of b. If both x and y are solution then z=ax + (1-a)y is also a solution for any real a.  <br>
Let v<sub>1</sub>, v<sub>2</sub>, ..,v<sub>r</sub> be vectors in R<sup>n</sup>. A linear combination of these vectors is any expression of the form k<sub>1</sub>v<sub>1</sub> + k<sub>2</sub>v<sub>2</sub> + ... + k<sub>r</sub>v<sub>r</sub> where coefficients k<sub>1</sub>, k<sub>2</sub>, ..,k<sub>r</sub> are scalars. <br>
The set of all linear combinations of collections of all vectors v<sub>1</sub>, v<sub>2</sub>, ..,v<sub>r</sub> from R<sup>n</sup> is called span of {v<sub>1</sub>, v<sub>2</sub>, ..,v<sub>r</sub>}. This particular span is known as column, space or range of V. If V=span{v<sub>1</sub>, v<sub>2</sub>, ..,v<sub>r</sub>} then V is set to be spanned by v<sub>1</sub>, v<sub>2</sub>, ..,v<sub>r</sub>. <br>
A set of vectors is linearly independent if no vector in the set is a linear combination of other vectors. <br>
When m=n in a matrix, then the matrix is called as a **square matrix**. A square matrix with linearly dependent column is called as **singular matrix**.

**Norms**

In machine learning we use a function called **norm** to measure the size of vectors. 

Let p ≥ 1 be a real number. L<sup>p</sup> norm is given by:
 \begin{align} \left\| \mathbf{x} \right\| _p := \bigg( \sum_{i=1}^n \left| x_i \right| ^p \bigg) ^{1/p}. \end{align} 

when p=2 it is called as Eucledian norm. A norm is any function *_f_* that satisfies: <br>
*_f_*(x+y) ≤ *_f_*(x) + *_f_*(y) <br>
*_f_*(x)=0 => x=0 <br>
*_f_*(αx) = |α|*_f_*(x) <br>

**Special Kinds of Matrices and Vectors**

1. Diagonal matrices consist mostly of zeros and have non zero entries along main diagonal.

3. A unit vector has a unit norm. 
4. Orthonormal vectors are those which have unit norm and are orthogonal.

**Eigendecomposition**

Eigendecomposition is a kind of matrix decomposition in which we decompose matrix into set of eigenvalues and eigenvectors. An eigenvector of a square matrix A is a nonzero vector v such that multiplication by matrix A only alters scale of v.

**Singular Value Decomposition**
The singular value decomposition provides a wat to factorize a matrix into singular vectors and values. The singular value decomposition of a matrix is a sort of change of coordinates that makes the matrix simple, a generalization of diagonalization.

**The Moore-Penrose Pseudoinverse**

The pseudoinverse are limits:
\begin{align}A^+ = \lim_{\delta \searrow 0} (A^* A + \delta I)^{-1} A^*
          = \lim_{\delta \searrow 0} A^* (A A^* + \delta I)^{-1}\end{align}

The following three properties of trace operator:

# B. Probability Theory

Probability is the measure of likelihood that an event will occur.

**Why probability?**
Machine learning must deal with uncertain and sometimes stochastic quantities, too. Probability theory provides a set of formal rules for determining the likelihood of a proposition being true given the other. 

**Random variables**
A random variable is a variable that can have any different value. 

**Probability Distribution**
A probability distribution is a distribution of likelihood of a random variable to take on each of its possible states. 

Discrete variables and probability mass function: 
Discrete variables are those that can only take finite number of values.
<br>
The probability mass function maps from a state of random variable to probability of it taking on that state. If a probability mass function acts on many variable at the same time it is called as joint probability distribution. 

**Continuous Variables and Probability Density Function**

Probability density function is a function of continuous random variable, whose integral across an interval gives the probability that the value of the variable lies within the same interval.


Probability of x lying in the interval a to b is given by: 

\begin{align} \Pr [a \le X \le b] = \int_a^b f_X(x) \, dx \end{align}

**Marginal Probability**
Marginal distribution of a subset of a collection of random variables is the probability distribution of the variables contained in the subset. 

Sum rule: 
\begin{align} \Pr(X=x) = \sum_y \Pr(X=x,Y=y) \end{align}

For continuous variables, integration is done as: 
 
 \begin{align} p_X(x) = \int_y p_{X,Y}(x,y) \, \mathrm{d}y  \end{align}


**Conditional Probability**

Conditional probability is the probability of an event given that other event has already occured. 


Chain rule:

\begin{align}\mathrm  P(X_n, \ldots , X_1)  = \mathrm P(X_n | X_{n-1}, \ldots , X_1) \cdot\mathrm P(X_{n-1}, \ldots , X_1)\end{align}


On repeating, we get: 

\begin{align}  P\left(\bigcap_{k=1}^n X_k\right)  = \prod_{k=1}^n  \mathrm P\left(X_k \,\Bigg|\, \bigcap_{j=1}^{k-1} X_j\right) \end{align}




**Independence and Conditional Independence**
Two random variables are independent if their probability distribution can be expressed as a product of two factors involving x and y independent of each other.

\begin{align} \mathrm{P}(A \cap B) = \mathrm{P}(A)\mathrm{P}(B) \end{align}


 
A and B are conditionally independent given C, if and only if : 

\begin{align} \Pr(A \cap B \mid C) = \Pr(A \mid C)\Pr(B \mid C) \end{align}

**Expectation, Variance and Covariance**
The expected value of X, where X is a discrete random variable, is a weighted average of the possible values that X can take, each value being weighted according to the probability of that event occurring.

Discrete variables: 
\begin{align}\operatorname{E}[X] =\sum_{i=1}^k x_i\,p_i=x_1p_1 + x_2p_2 + \cdots + x_kp_k \end{align}

<br>
Continuous variables:
\begin{align} \operatorname{E}[X] = \int_{\R} x f(x)\, dx \end{align}
<br>

The variance of a random variable tells us something about the spread of the possible values of the variable.

\begin{align} \operatorname{Var}(X) = \operatorname{E}\left[(X - \mu)^2 \right] \end{align}

Standard deviation: SD is a measure that is used to quantify the amount of variation of a set of data values. 


Covariance: Covariance tell us how much two values are linearly related to each other.

\begin{align}\operatorname{cov}(X,Y) = \operatorname{E}{\big[(X - \operatorname{E}[X])(Y - \operatorname{E}[Y])\big]}\end{align}

<b>Observation:</b>  
Outlier Treatment using 1.5 IQR rule  
Rule: Outliers here are defined as observations that fall below Q1 − 1.5 IQR or above Q3 + 1.5 IQR.

```python
def benchR_func(q25,q75):
    return (q75 + 1.5*(q75-q25))

def benchL_func(q25,q75):
    return (q25 - 1.5*(q75-q25))

def out_rem(x,benchL,benchR):
    if x>benchR:
        return benchR
    
    elif x<benchL:
        return benchL
    else :
        return x
```

```python
labels = ['bedrooms','sqft_living','floors','sqft_above','sqft_basement','sqft_living15','sqft_lot15']
for label in labels :
    q25,q75 = q75, q25 = np.percentile(df_homeprice[label], [75,25])
    benchL = benchL_func(q25,q75)
    benchR = benchR_func(q25,q75)
    df_homeprice[label]= df_homeprice[label].apply(lambda row: out_rem(row,benchL,benchR))
```

```python
# Box plot after outlier treatment :
df_homeprice.drop(['id','date','waterfront','view','grade','yr_built','zipcode','lat','long','yr_renovated'],axis = 1).plot(kind='box', subplots=True, layout=(4,3), sharex=False, sharey=False,
        figsize=(9,9),title='Box Plots after outlier removal from some columns')
plt.tight_layout()
```
